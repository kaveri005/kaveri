```python
Slovakia_case.groupby(['new_cases']).size().reset_index(name='Count').rename(columns={'new_cases':'new_cases'})
#finding unique values in new_cases column
```

```python
Slovakia_case.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,7))
x=Slovakia_case['new_cases']
plt.hist(x,bins=20,color='blue')
plt.title('new_cases')
plt.xlabel('new_cases')
plt.ylabel('Count')
```

```python
sns.lmplot(x='date',y='new_cases',fit_reg=False,data=Slovakia_case)
```

```python
Slovakia_case.corr()
```

```python
#Totalcases per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='date',y='new_cases',data=Slovakia_case)
plt.show()
```

```python
#looking a dataframefor last 5 days
Slovakia_last_five_days=Slovakia_case.tail()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='date',y='new_cases',data=Slovakia_last_five_days)
plt.show()
```

```python
last_day_cases=Slovakia_case[Slovakia_case['date']=='2020-05-24']#lates cases
last_day_cases
```

```python
last_day_cases=covid[covid['date']=='2020-05-24']#latest cases
last_day_cases
```

```python
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)#sorting data with total cases
max_cases_country
```

```python
max_cases_country[1:10]
#ranking the countries according to the cases registered
```

```python
sns.barplot(x='location',y='new_cases',data=max_cases_country[1:10],hue='location')
plt.show()
```
